```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
##plotting using matplotlib
##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
# plt.savefig('test.png')
```

```python
y = x*x
```

```python
##plt plot
plt.plot(x,y,'r*-')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
##creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

```python
x = np.arange(1,11)
y = 3*x+5
plt.plot(x,y,'b+')
```

```python
#compute the x and y coordinates for a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("Sine wave")

plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('Sin')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cos')
```

```python
## bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

```python
##histograms
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=50)
```

```python
##box plot
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert = True,patch_artist=True)
```

```python
##pie chart
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0.1,0)

plt.pie(sizes,explode=explode, labels=labels, colors=colors,
        autopct ='%1.1f%%',shadow=False)
plt.axis('Equal')
```
