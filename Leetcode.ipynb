## Drop Useless Colums


```python
df.drop(['id','page_number','topic_tags','problem_URL','solution_URL','problem_description','no_similar_questions','similar_questions','solution','discussion_count'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['difficulty'].unique()
```

```python
df['difficulty'] = df['difficulty'].map({'Easy':1,'Medium':2,'Hard':3})
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['accepted'].mode()[0]
```

```python
df['accepted'] = df['accepted'].fillna(df['accepted'].mode()[0])
df.isnull().sum()
```

```python
df['submission'] = df['submission'].fillna(df['submission'].mode()[0])
df['likes'] = df['likes'].fillna(df['likes'].mode()[0])
df['dislikes'] = df['dislikes'].fillna(df['dislikes'].mode()[0])
df.isnull().sum()
```

## Vizualization
